```python
import os
import cv2
import numpy as np
import datetime
import json
from PIL import Image
```

```python
flist = os.listdir('../nftassets/Random')
flist = sorted(flist)

pdir = '../nftassets/Random/'
```

```python
json_list = []
for fname in flist:
    if fname[-4:] == 'json':
        with open(pdir + fname) as fp:
            json_data = json.load(fp)
        #print(json_data)
        json_list.append(json_data)
```

```python
tdir = '../nftassets/renamed/'
```

```python
import tqdm
```

```python
for json_data in tqdm.tqdm(json_list):
    traits = {trait['trait_type']:trait['value'] for trait in json_data['attributes']}
    
    animals = ['fox', 'meerkat', 'dog', 'penguin'] #B
    item_ie = ['game', 'balloon', 'mic', 'newspaper'] #E
    item_sn = ['toilet_paper','helmet', 'crown', 'cap'] #E
    item_ft = ['blue', 'green', 'yellow', 'red'] #D
    item_jp = ['home', 'office', 'park', 'beach'] #C
    
    ai = int(traits['B'][1])-1
    ie = int(traits['A'][1])-1
    sn = int(traits['E'][1])-1
    ft = int(traits['D'][1])-1
    jp = int(traits['C'][1])-1
    
    imgpath = json_data['image']
    newname = f"{animals[ai]}-{item_ie[ie]}-{item_sn[sn]}-{item_ft[ft]}-{item_jp[jp]}.png"
    
    img = Image.open(pdir + imgpath)
    img.save(tdir + newname)
```
